```python
colunas_terminadas_em_0 = [coluna for coluna in df.columns if coluna.endswith('0')]
colunas_terminadas_em_1 = [coluna for coluna in df.columns if coluna.endswith('1')]
colunas_terminadas_em_2 = [coluna for coluna in df.columns if coluna.endswith('2')]
df_2020 = df[colunas_terminadas_em_0]
df_2021 = df[colunas_terminadas_em_1]
df_2022 = df[colunas_terminadas_em_2]
```

```python
def troca_valor(a, b, column):
  for indice, valor in column.items():
      if valor == a:
          df_2020_filtrado.loc[indice, b] = 1
      else:
          df_2020_filtrado.loc[indice, b] = 0
```

```python
troca_valor('Escola Pública','INSTITUICAO_ENSINO_ALUNO_2020', df_2020_filtrado.INSTITUICAO_ENSINO_ALUNO_2020)
```
